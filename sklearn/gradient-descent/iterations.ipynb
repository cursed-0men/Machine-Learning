# Gradient descent

```python
import numpy as np
import pandas as pd
```

```python
# datasets
from sklearn.datasets import make_regression
x,y = make_regression(n_samples = 4, n_features = 1, n_informative = 1, n_targets = 1, noise = 80, random_state = 13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# to find slope(m) value
# using least squares
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

```python
# slope
lr.coef_
```

```python
# intercept
lr.intercept_
```

```python
# line of regression
plt.scatter(x,y)
plt.plot(x, lr.predict(x), color = 'green', label = 'least squares BFL')
plt.legend()
```

### Applying gradient Descent
> having slope = 78.35063668

> assuming a random value of b (lets say, 0)

```python
# drawing that line with b = 0
y_pred = ((lr.coef_ * x) + 0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x, lr.predict(x), color = 'green', label = 'least squares BFL')
plt.plot(x, y_pred, color = 'red', label = 'at b = 0')
plt.legend()
```

> iteration 1

```python
# with gradient descent red line will move toward, green line
m = 78.35063668
b = 0
n = 0.1   # learing rate typical value is 0.01
```

```python
loss_slope = (-2/4) * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
step_size = n * loss_slope
step_size
```

```python
# new intercept
b = b - step_size
b
```

```python
y_pred1 =(m*x + b).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x, lr.predict(x), color = 'green', label = 'least squares BFL')
plt.plot(x, y_pred, color = 'red', label = 'at b = 0')
plt.plot(x, y_pred1, color = 'orange', label = 'at b = {}'.format(b))
plt.legend()
```

> iteration 2

```python
# loss slope
loss_slope = (-2/4) * np.sum(y - m*x.ravel() -b)
loss_slope
```

```python
step_size = n * loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = (m*x + b).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x, lr.predict(x), color = 'green', label = 'least squares BFL')
plt.plot(x, y_pred, color = 'red', label = 'at b = 0')
plt.plot(x, y_pred1, color = 'orange', label = 'at b = {}'.format(b))
plt.plot(x, y_pred2, color = 'blue', label = 'at b = {}'.format(b))
plt.legend()
```

> iterations in loop

```python
b = -100
m = 78.35063668
n = 0.1     # learning rate


epochs = 10     # iterations

plt.figure(figsize = (20,10))
for i in range(epochs):
    loss_slope = (-2/4) * np.sum(y - m*x.ravel() - b)
    b = b - (loss_slope * n)

    y_pred = m*x + b
    plt.plot(x,y_pred)

plt.scatter(x,y)
```
