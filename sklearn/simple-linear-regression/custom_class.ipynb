## Customized Linear Regression
- Writing class
- using <span style = 'color:orange'>Least Squares method</span>


```python
class dhyeys_linear_regression:
    
    def __init__(self):
        self.m = 0
        self.b = 0

    def fit(self, x_train, y_train):
        num = 0
        den = 0

        for i in range(x_train.shape[0]):
            num += (x_train[i] - x_train.mean())*(y_train[i] - y_train.mean())
            den += (x_train[i] - x_train.mean())**2

        self.m = num/den
        self.b = y_train.mean() - (self.m * x_train.mean())
        print('Model Trained...')
        print('slope:',self.m)
        print('intercept:',self.b)
    
    def predict(self, x_test):
        print('Predictions: \n')
        return self.m * x_test + self.b
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
df.head()
```

```python
# feature(x) and target(y)
x = df.iloc[:, 0].values
y = df.iloc[:,1].values
```

```python
# true data points
plt.scatter(x,y,s = 70, color = 'darkorange', label = 'True data points')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.legend()
plt.show()
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
# custom Linear Regression
lr = dhyeys_linear_regression()
lr.fit(x_train, y_train)
```

```python
y_test
```

```python
# predictions
lr.predict(x_test)
```
