# Linear Regression : Gradient Descent

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test.csv')
df.head()
```

```python
x = df['house_size'].values
y = df['price_usd'].values

# standardizing the data
x = (x - np.mean(x))/np.std(x)
y = (y - np.mean(y))/np.std(y)
```

```python
x,y
```

```python
# plotting existing data
plt.scatter(x,y,color = 'lightgreen', edgecolor = 'green',s = 80,label = 'actual data')
plt.xlabel('house size')
plt.ylabel('price usd')
plt.title('house size vs price usd')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# hyper parameters
m = 0
b = 0

steps = 1000
lr = 0.01
n = len(x)

residual_list = [] # to track error on each step
```

```python
for i in range(steps):
  y_pred = m*x + b
  residual = y - y_pred

  loss_mse = (1/n)* np.sum((residual)**2)
  residual_list.append(loss_mse)

  # gradients
  dm = (-2/n)*np.sum(x * residual)
  db = (-2/n)*np.sum(residual)

  # updating the weights
  m = m - (lr * dm)
  b = b - (lr * db)

  print(f"step {i}, slope:{m}, intercept:{b}, loss:{loss_mse}")
```

```python
print(' final slope :',round(m,4))
print(' final intercept :',round(b,4))
```

```python
plt.scatter(x,y,color = 'lightgreen', edgecolor = 'green', s = 80,label = 'Actual Data')
plt.plot(x,y_pred,color = 'blue',label = 'fitted line')
plt.xlabel('house size')
plt.ylabel('price usd')
plt.title('house size vs price usd')
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_pred = m*x + b
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y,y_pred)
mae
```

```python
# loss curve
plt.figure(figsize = (10,10))
plt.plot(range(steps),residual_list,color = 'green')
plt.xlabel('Steps')
plt.ylabel('Loss (MSE)')
plt.title('Loss Curve')
plt.grid(True)
plt.show()
```
